```python
import pandas as pd
data = {
    'id': [1, 2, 3, 4],
    'name': ['Joe', 'Henry', 'Sam', 'Max'],
    'salary': [70000, 80000, 60000, 90000],
    'managerId': [3, 4, None, None]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
salaries = df.set_index('id').salary
```

```python
salaries
```

```python
df['mgr_salary'] = df['managerId'].map(salaries)
```

```python
df
```

```python
print(df[df["salary"] > df['mgr_salary']]['name'])
```
