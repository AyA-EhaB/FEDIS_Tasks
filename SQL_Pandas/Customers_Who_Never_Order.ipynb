```python
import pandas as pd
```

```python
C = {'id':[1,2,3,4],'name':['Joe','Henry','Sam','Max']}
C
```

```python
customrs = pd.DataFrame(C)
```

```python
customrs
```

```python
order = pd.DataFrame({'id':[1,2],'customerId':[3,1]})
order
```

```python
merged = pd.merge(customrs,order,left_on='id',right_on='customerId',how='left')
merged
```

```python
never_ordered = merged[merged['customerId'].isna()]
never_ordered['name']
```
